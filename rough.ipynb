```python
import pandas as pd
import shapely as shp
import geopandas as gpd
```

```python
df = pd.read_parquet('FimaNfipClaims.parquet')
```

```python
list(df['nfipCommunityName'].unique())
```

```python
df[['latitude', 'longitude']].fillna(0, inplace=True)
df['geometry'] = df.apply(lambda row: shp.Point(row['longitude'], row['latitude']), axis=1)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs = 'EPSG:4326')
gdf.head()
```
